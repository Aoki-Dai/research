## wiki40b_ja(wikipedia)のコーパスをロード

```python
from datasets import load_dataset

ds = load_dataset("fujiki/wiki40b_ja")
```

```python
# データセットの構造を確認
print("データセット情報:")
print(ds)
print("\n利用可能なスプリット:")
print(ds.keys())
print("\ntrain スプリットのサンプル数:")
print(len(ds['train']))
print("\nデータの最初の1件:")
print(ds['train'][0])
```

```python
import pandas as pd

# データをDataFrameに変換（サンプリング）
sample_size = min(10000, len(ds['train']))
sample_data = ds['train'].select(range(sample_size))
df = pd.DataFrame(sample_data)
```

```python
# DataFrameの基本情報
print("DataFrameの形状:", df.shape)
print("\nカラム名:")
print(df.columns.tolist())
print("\n最初の3件:")
df.head(3)
```

```python
# サンプルテキストの表示
print("=" * 80)
print("サンプルテキスト（最初の3件）")
print("=" * 80)
for i in range(min(3, len(df))):
    print(f"\n【記事 {i+1}】")
    if 'title' in df.columns:
        print(f"タイトル: {df.iloc[i]['title']}")
    if 'text' in df.columns:
        text = df.iloc[i]['text']
        # 最初の300文字のみ表示
        print(f"本文（抜粋）: {text[:300]}...")
    print("-" * 80)
```
